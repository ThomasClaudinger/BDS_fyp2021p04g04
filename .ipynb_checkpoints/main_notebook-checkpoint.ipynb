### Remove punctuations from txt files

```python
def del_punctuations(words):

    punctuation_table = str.maketrans('', '', string.punctuation)

    words = [word.translate(punctuation_table).lower() for word in words]
    return words

cleaned_val_words = del_punctuations(words_val_txt[:100])
cleaned_train_words = del_punctuations(words_train_txt[:100])
cleaned_test_words = del_punctuations(words_test_txt[:100])

print('val_text.txt:\n',cleaned_val_words,'\n')

print('\nval_train.txt:\n',cleaned_train_words, '\n')

print('\nval_test.txt:\n',cleaned_test_words,'\n')
```

### Number of words in each tokenization variables

```python
from nltk.tokenize import word_tokenize

def tokenization(words, text):

    tokens = word_tokenize(text[1:])
    words = [token.lower() for token in tokens if token.isalpha()]
    
    return words

token_val = tokenization(cleaned_val_words, text_val)
token_train = tokenization(cleaned_train_words, text_train)
token_test = tokenization(cleaned_test_words, text_test)

print(f'Number of words in tokenization for val_text: {len(token_val)}')
print(f'Number of words in tokenization for val_train: {len(token_train)}')
print(f'Number of words in tokenization for val_test: {len(token_test)}')
```
